```python
# First generating two X columns and one Y column using pandas and numpy
import pandas as pd
import numpy as np

X = pd.DataFrame(np.random.randint(0,1000,size=(1000,2)), columns=['x1','x2'])
Y = pd.DataFrame(np.random.randint(2,size=(1000,1)), columns=['y'])
```

```python
print(X)
```

```python
print(Y)
```

```python
# Generating three possible coefficents and weights (b1,b2,b) of the equation y = b + b1x1 + b2x2
# I used random number gerator to guess possible vaulues for b,b1,b2. In this I may be wrong.
# Also calculatig z = b + b1x1 + b2x2 for each value of x1 and x2
Z = []
for x in range(len(X)):
    b = np.random.randint(-100,100)
    b1 = np.random.randint(-100,100)
    b2 = np.random.randint(-100,100)
        
    B = (Y.y[x] - (b + b1*X.x1[x] + b2*X.x2[x]))
    B1 = ((Y.y[x] - (b + b1*X.x1[x] + b2*X.x2[x]))*2*X.x1[x])/len(X)
    B2 = ((Y.y[x] - (b + b1*X.x1[x] + b2*X.x2[x]))*2*X.x2[x])/len(X)
    z = B + B1*X.x1[x] + B2*X.x2[x]
    Z.append(z)
        
```

```python
print(Z)
```

```python
# calculating the probability using sigmoid function
prob = []
for x in range(len(Z)):
    z = Z[x]
    t = 2.71828**z
    sig = t/(1 + t)
    prob.append(sig)
```

```python
print(prob)
```

```python
#Checking if the predicted probability is greater than equal to or less than 0.5 and appending 1 and 0 
# in a new list
prediction =[]
for x in range(0,len(prob)):
    p = prob[x]
    if (p>=0.5):
        prediction.append(1)
    else:
        prediction.append(0)
```

```python
print(prediction)
```

```python
#Calculating true posistive and true negative using below logic
TN = 0
TP = 0
for x in range(0,len(Y)):
    if Y.y[x] == 1 and prediction[x] == 1:
        TP +=1
for x in range(0,len(Y)):
    if Y.y[x] == 0 and prediction[x] == 0:
        TN +=1
```

```python
print(TP,TN)
```

```python
# calculating accuracy of the model
accuracy = ((TP + TN)/len(Y))*100
print(accuracy)
```
